# (-5) Incorrect Logic. Keep track of the most frequent user moves and then decide

```python
"""
Problem 1:
    - Approach:
        Problem will be split up into 3 functions consisting of...
        1. User_Option: Takes user input and validates it
        2. Computer_Option: Plays option that beats users most common input
        3. While loop to start the game and keep it going
        
"""
from pandas import Series, DataFrame
import pandas as pd
import random

games_played = 0
user_wins = 0
quit = "no"

def user_option():
    '''
    Input: User chooses rock, paper, or scissors
    Output: Validated user's choice
    '''
    user_choice = input("\nChoose Rock, Paper, or Scissors: ")
    if user_choice in ["Rock", "rock", "R", "r"]:
        user_choice = "Rock"
        
    elif user_choice in ["Paper", "paper", "P", "p"]:
        user_choice = "Paper"
        
    elif user_choice in ["Scissors", "Scissor", "scissors", "scissor", "S", "s"]:
        user_choice = "Scissors"
    
    else:
        print("\nI did not understand that. Please try again. ")
        pass
        #user_option()
    return user_choice

def comp_option():
    '''
    Output: Computer's choice
    '''
    comp_choice = random.randint(0,2)
    if comp_choice == 0:
        comp_choice = "Rock"
    elif comp_choice == 1:
        comp_choice = "Paper"
    else:
        comp_choice = "Scissors"
    return comp_choice


while quit == "no":
    '''
    Describes the rules of the game and takes user input to either end game or keep playing
    Output: Results of the game
    '''
    comp_choice = comp_option()
    user_choice = user_option()

    if comp_choice == "Rock":
        if user_choice == "Rock":
            print("You both chose rock. It was a tie.")
            games_played += 1
        elif user_choice == "Paper":
            print("Your paper beat their rock. You won!")
            games_played += 1
            user_wins += 1
        elif user_choice == "Scissors":
            print("Your scissors lost to their rock. You lost.")
            games_played += 1

    elif comp_choice == "Paper":
        if user_choice == "Paper":
            print("You both chose paper. It was a tie.")
            games_played += 1
        elif user_choice == "Rock":
            print("Your rock lost to their paper. You lost.")
            games_played += 1
        elif user_choice == "Scissors":
            print("Your scissors beat their paper. You won!")
            games_played += 1
            user_wins += 1

    elif comp_choice == "Scissors":
        if user_choice =="Scissors":
            print("You both chose scissors. It was a tie.")
            games_played += 1
        elif user_choice == "Rock":
            print("Your rock beat their scissors. You won!")
            games_played += 1
            user_wins += 1
        elif user_choice == "Paper":
            print("Your paper lost to their scissors. You lost.")
            games_played += 1

    user_choice = input("\nDo you want to keep playing? If so, type yes: ")
    if user_choice in ["Yes", "yes", "Y", "y"]:
        pass
    else:
        print("\nGames played: " + str(games_played))
        print("Games won: " + str(user_wins))
        quit = "yes"
```

```python
'''
For this problem, I thought it would be best to cycle through each of the lines in the song skeleton, and search through
the possible endings based on the given line's rhyming scheme and beats needed to complete the line, which could both
be calculated from the most recent complete line. I first started off by creating empty lists to add the information
from the endings.txt file to. My first loop cycles through each line of the endings file and splits all the necessary
variables into their respective lists. I then made two series(one for rhyming scheme and one for beats per ending), both
indexed with the ending words because they were all unique and could link the other two variables together. Now I wanted 
to cycle through all the lines of the song skeleton in order, and I needed different conditional actions to take for each
one, so I decided to implement some nested if/else statements. The first type of line, where the ending is given, sets
the variables lineRhyme and lineBeats, representing the rhyming scheme and total beats per line. Since all the unknown
lines will follow these rules, the variables were designed to only be updated every time a completed line is appears.
The next type of line is that with an unknown ending. For these, I filtered down the two series created from the endings
file with masks based on the known rhyming scheme and remaining beats needed. Since every combination of those variables
yielded a unique ending, I could simply take the first Index of the new series and add it onto the lyric I was trying to
complete. The last type of line has no ending, and I would just keep the given lyrics. After each one of these conditions 
and concatenations, I would add the completed lines to a list called 'song', and then the final step was to loop through 
each value from that list and join them together into one big string to give nice, presentable song lyrics.
'''
import pandas as pd
import numpy as np

fpe = open('endings.txt', 'r')
fps = open('skeleton_SamuraiShowdown.txt','r') #read provided files

ends = []
rhymes = [] #initialize lists for ending variables
beats = []

song = []

endline = fpe.readline()

while endline != '': #loop through each ending line and break up string into variables
    endline = endline.rstrip()
    end, rhyme, beatnum = endline.split('::')
    
    ends.append(end)
    rhymes.append(rhyme) #append variables to their respective lists
    beats.append(beatnum)
    
    endline = fpe.readline()

rhyme_scheme = pd.Series(rhymes, index = ends) #create the two series to reference ending variables
end_beats = pd.Series(beats, index = ends)


line = fps.readline()

lineBeats = 0
lineRhyme = '' #initialize holding/calculation variables
needBeats = 0

while line != '':
    rhyme_scheme = pd.Series(rhymes, index = ends) #have to reset the series before they get filtered down within the loops
    end_beats = pd.Series(beats, index = ends)
    
    line = line.rstrip()
    lyric, beatnum, end1 = line.split('::') #break up skeleton lines into lyric, beat, and ending variables
    
    if end1 != '': #for all non-null ending values
        
        if end1 != 'XXX': #for all given endings
            lineRhyme = rhyme_scheme[end1]
            lineBeats = int(beatnum) + int(end_beats[end1]) #set the rhyme and beat variables until the next complete line
            
            line = lyric + ' ' + end1
            song.append(line)
            
        else: #for all unknown or 'XXX' endings
            mask = (rhyme_scheme == lineRhyme)
            end_beats = end_beats[mask]     #filter series based of all rhyming matches
            
            needBeats = lineBeats - int(beatnum)
            mask1 = (end_beats == str(needBeats))
            end_beats = end_beats[mask1]   #further filter series based of the remaining needed beats
            
            end2 = end_beats.index[0]   #extract the correct ending
                            
            line = lyric + ' ' + end2
            song.append(line)
            
    else:  #for all null ending values
        line = lyric
        song.append(line)

    line = fps.readline() #move onto the next line of the skeleton
    
songText = ''
for i in song:  #loop through list of song lines and join them into one string
    songText = songText + i + '\n'

print(songText)
```
