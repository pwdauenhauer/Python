```python
import pandas as pd
import numpy as np

tally = pd.Series({'rock':0, 'paper':0, 'scissors':0})
userChoice = ''
gameCount = 0
winCount = 0

def main(series, gameC, win):
    userChoice = input('Choose rock, paper, or scissors: ')
    
    if userChoice in['rock', 'paper', 'scissors', 'q']:
        game(userChoice, series, gameC, win)
        
    else:
        print("That's not a choice dummy, try again")
        main(tally)

def compChoice(series):
    predict = series.idxmax()
    choice = ''
    if predict == 'rock':
        choice = 'paper'
    elif predict == 'paper':
        choice = 'scissors'
    elif predict == 'scissors':
        choice = 'rock'
    return(choice)
    
def game(uChoice, series, gameC, win):
    sysMove = compChoice(series)
    print('My move is: ',sysMove)
    gameC = gameC + 1
    if uChoice == 'rock':
        if sysMove =='rock':
            print('A worthy opponent, draw')
        elif sysMove == 'paper':
            print('lmao u lost')
        elif sysMove =='scissors':
            print('dam bro u pretty good')
            win = win + 1
                    
    elif uChoice == 'paper':
        if sysMove =='paper':
            print('A worthy opponent, draw')
        elif sysMove == 'scissors':
            print('lmao u lost')
        elif sysMove =='rock':
            print('dam bro u pretty good')
            win = win + 1
    
    elif uChoice == 'scissors':
        if sysMove =='scissors':
            print('A worthy opponent, draw')
        elif sysMove == 'rock':
            print('lmao u lost')
        elif sysMove =='paper':
            print('dam bro u pretty good')
            win = win + 1
        
    else:
        print('k bye')
    
    print('You have won ',str(win), ' of ', str(gameC), ' games')
    series.loc[uChoice] = series.loc[uChoice] + 1
    print(series)
    main(series, gameC, win)
    
main(tally, gameCount, winCount)
```
